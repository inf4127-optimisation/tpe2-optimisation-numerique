# ExpÃ©rimentations : Fonction de Himmelblau avec Pas Optimal

## 1.1 DÃ©finition de la Fonction

La **fonction de Himmelblau** est dÃ©finie par :

$$f(x, y) = (x^2 + y - 11)^2 + (x + y^2 - 7)^2$$

Cette fonction est cÃ©lÃ¨bre en optimisation car elle possÃ¨de **quatre minima locaux** de valeur nulle.

### CaractÃ©ristiques principales :
- **4 minima locaux** avec $f(x,y) = 0$
- **Points critiques** : solutions du systÃ¨me d'Ã©quations
- **Surface complexe** avec plusieurs bassins d'attraction

## 1.2 Calcul du Gradient

Le gradient de la fonction est :

$$\nabla f(x, y) = \begin{pmatrix}
\frac{\partial f}{\partial x} \\
\frac{\partial f}{\partial y}
\end{pmatrix} = \begin{pmatrix}
4x(x^2 + y - 11) + 2(x + y^2 - 7) \\
2(x^2 + y - 11) + 4y(x + y^2 - 7)
\end{pmatrix}$$

````python
from sympy import re, im

def solutions_reelles_numeriques(solutions):
    solutions_reelles = []
    for sol in solutions:
        x_val = sol[x]
        y_val = sol[y]
        
        # Ã‰valuer numÃ©riquement
        x_num = x_val.evalf()
        y_num = y_val.evalf()
        
        # VÃ©rifier si c'est rÃ©el
        if abs(im(x_num)) < 1e-10 and abs(im(y_num)) < 1e-10:
            solutions_reelles.append((float(re(x_num)), float(re(y_num))))
    
    return solutions_reelles

# Utilisation
solutions_num = solutions_reelles_numeriques(solutions)
print("Solutions rÃ©elles numÃ©riques:", solutions_num)

point_min = []
for i in range(len(solutions_num)):
    # CORRECTION : CrÃ©er un dictionnaire pour subs
    point_dict = {x: solutions_num[i][0], y: solutions_num[i][1]}
    
    # Ã‰valuer la Hessienne au point
    Hf_point = Hf.subs(point_dict)
    
    # Calculer le dÃ©terminant
    det_H = Hf_point.det()
    
    print(f"Point {solutions_num[i]}: det(H) = {det_H}")
    
    if det_H > 0:
        point_min.append(solutions_num[i])
        print('  â†’ Minimum local')
    elif det_H < 0:
        print('  â†’ Point selle')
    else:
        print('  â†’ Cas dÃ©gÃ©nÃ©rÃ© (dÃ©terminant = 0)')
````

````python
# Si vous voulez une fonction qui retourne directement grx et gry
def gradient_complet(graf_list, point_x_y):
    print('point', point_x_y)
    grx_func = sp.lambdify((x, y), graf_list[0], 'numpy')
    gry_func = sp.lambdify((x, y), graf_list[1], 'numpy')
    grx = grx_func(point_x_y[0], point_x_y[1])
    gry = gry_func(point_x_y[0], point_x_y[1])
    return grx, gry
````

````python
def pas_fixed():
    pas_optimal = 0.001
    return pas_optimal
````

````python
def pas_optimal(optimal, point_x_y):
    print('point', point_x_y)
    # VÃ©rifier si optimal est une liste/tuple avec plusieurs Ã©lÃ©ments ou une seule valeur
    if hasattr(optimal, '__len__') and len(optimal) > 1:
        pas_x = optimal[0].subs({x: float(point_x_y[0]), y: float(point_x_y[1])})
        pas_y = optimal[1].subs({x: float(point_x_y[0]), y: float(point_x_y[1])})
        return float(pas_x.evalf()), float(pas_y.evalf())
    else:
        # Si optimal est une seule valeur
        pas = optimal.subs({x: float(point_x_y[0]), y: float(point_x_y[1])})
        return float(pas.evalf())
````

````python
def descente_gradient(grad, optimal, point_xy, maxIter, tol):
    history_grad = []
    history_f = []
    history_pas = []
    history_points = [point_xy]  # Stocker tous les points
    current_point = point_xy
    
    for i in range(maxIter):
        print(f"\n--- ItÃ©ration {i+1} ---")
        print(f"Point actuel: {current_point}")
        
        # Calcul du gradient
        gx, gy = gradient_complet(grad, current_point)
        history_grad.append((gx, gy))
        print(f"Gradient: ({gx}, {gy})")
        
        # Calcul du pas optimal
        pas_result = pas_fixed()
        print(f"Pas optimal: {pas_result}")
        history_pas.append(pas_result)
        
        # Mise Ã  jour du point (conserver comme tuple)
        if isinstance(pas_result, tuple):
            # Si pas_result est un tuple (pas_x, pas_y)
            new_x = current_point[0] - pas_result[0] * gx
            new_y = current_point[1] - pas_result[1] * gy
        else:
            # Si pas_result est une seule valeur
            new_x = current_point[0] - pas_result * gx
            new_y = current_point[1] - pas_result * gy
        
        current_point = (new_x, new_y)  # Conserver comme tuple
        history_points.append(current_point)
        
    return history_points, history_grad, history_pas
````

````python
# Tracer aussi la surface avec le chemin de descente
def tracer_surface_avec_chemin(history_points, f):
    # CrÃ©er un grid pour la surface
    x_min = min(p[0] for p in history_points) - 1
    x_max = max(p[0] for p in history_points) + 1
    y_min = min(p[1] for p in history_points) - 1
    y_max = max(p[1] for p in history_points) + 1
    
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 50),
                         np.linspace(y_min, y_max, 50))
    
    # Convertir f en fonction numÃ©rique
    f_num = sp.lambdify((x, y), f, 'numpy')
    zz = f_num(xx, yy)
    
    # Tracer
    fig = plt.figure(figsize=(12, 5))
    
    # Surface 3D
    ax1 = fig.add_subplot(121, projection='3d')
    ax1.plot_surface(xx, yy, zz, alpha=0.6, cmap='viridis')
    
    # Tracer le chemin de descente sur la surface
    z_path = [f_num(p[0], p[1]) for p in history_points]
    ax1.plot([p[0] for p in history_points], 
             [p[1] for p in history_points], 
             z_path, 'ro-', linewidth=2, markersize=4)
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_zlabel('f(x,y)')
    ax1.set_title('Surface et chemin de descente')
    
    # Contour 2D
    ax2 = fig.add_subplot(122)
    contour = ax2.contour(xx, yy, zz, levels=20, alpha=0.6)
    ax2.clabel(contour, inline=True, fontsize=8)
    ax2.plot([p[0] for p in history_points], 
             [p[1] for p in history_points], 'ro-', markersize=4)
    ax2.plot(history_points[0][0], history_points[0][1], 'go', markersize=8, label='DÃ©part')
    ax2.plot(history_points[-1][0], history_points[-1][1], 'bo', markersize=8, label='Final')
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_title('Lignes de niveau et chemin')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()

# Tracer la surface avec le chemin
tracer_surface_avec_chemin(hp, f)
````
